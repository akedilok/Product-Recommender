# Data Cleaning

## Remove Nulls

```python
# Drop all rows with null values
all_data = all_data.dropna(how = 'any', axis = 0)
```

```python
# Look dataframe info to confirm no nulls remain
all_data.info()
```

## Remove Duplicates

```python
# Create a boolean column, 'duplicates', that will be True if row is a duplicate and False if not
# Duplicates will be determined using subset of columns entered
# Description not included in subset columns because there are errors present that results in duplicates being flagged as unique
# keep = 'first': marks instances after first as 'True'
all_data['duplicates'] = all_data.duplicated(subset = ['StockCode', 
                                                       'InvoiceNo', 
                                                       'InvoiceDate', 
                                                       'CustomerID', 
                                                       'Quantity', 
                                                       'UnitPrice'], keep = 'first')
```

```python
# Keep rows where duplicates column value was False (aka non-duplicate rows) and drop duplicates column
all_data = (all_data[all_data.duplicates == False].sort_values('InvoiceDate', ascending = True)
                                                 .drop(['duplicates'], axis = 1).reset_index(drop=True))
```

```python
# all_data info after duplicates were dropped
all_data.info()
```

## 'Description' Clean-up

**1. Drop rows were Description is 'Manual' and no product name.**  

```python
# Keep only rows where Description is not 'Manual'
all_data = all_data[all_data['Description'] != 'Manual'].reset_index(drop = True)
```

```python
# all_data info after Descriptions of 'Manual' were dropped
all_data.info()
```

**2. Remove spaces before and after descriptions and remove non-alphanumeric characters.**

```python
# Strip spaces before and after descriptions
all_data.Description = all_data.Description.apply(lambda x: x.strip())
```

```python
# Replace any non-alphanumeric character with a space
all_data.Description = all_data.Description.apply(lambda x: re.sub('[^0-9a-zA-Z\' ]+', ' ', x))
```

```python
# Replace multiple spaces with one space
all_data.Description = all_data.Description.apply(lambda x: re.sub(' +', ' ', x))
```

```python
# Visualize all descriptions
for x in all_data.Description.sort_values(ascending = True).unique():
    print(x)
```

**3. Replace Descriptions for StockCodes with multiple descriptions with a uniform description.**

```python
def count_desc(df):
    '''Function finds and counts StockCodes with more than one Description.  Function returns the number of
       StockCodes with duplicate Descriptions and a list of the codes.  
       Inputs:
         df = dataframe to reference
         code_count = counter to record number of StockCodes with > 1 Description
         mult_desc_codes = empty list initiated to record StockCodes with multiple descriptions
       Outputs:
         code_count and mult_desc_codes
     '''
    code_series = df.groupby(['StockCode']).Description.nunique().sort_values(ascending = False)
    code_count = 0
    mult_desc_codes = []
    
    # For each StockCode, increment code_count and record StockCode to list if number of unique Description > 1 
    for row in range(0,len(code_series)):
        if code_series.iloc[row] > 1:
            code_count += 1
            mult_desc_codes.append(code_series.index[row])
    return code_count, mult_desc_codes
```

```python
# Run count_desc() function and store outputs
code_count, mult_desc_codes = count_desc(all_data)
code_count, mult_desc_codes
```

```python
# Replace alternate descriptions with the first description used
for code in mult_desc_codes:
    descriptions = list(all_data[all_data.StockCode == code].Description.unique())
    master_desc = descriptions[0]
    
    if len(descriptions) > 1:
        # clears printed values
        clear_output(wait=True)
        
        # print current status
        print(f'\rCurrently replacing: {descriptions}\n\twith {master_desc}', end="\r", flush=True)
        
        # replace description
        for versions in descriptions[1:]:
            all_data.Description = all_data.Description.replace(versions, master_desc)
    else:
        continue
        
# check that no StockCodes have multiple descriptions
if all_data.groupby(['StockCode']).Description.nunique().sort_values(ascending = False).max() == 1:
    clear_output(wait=True)
    print(f'Finished!  All duplicate descriptions have been removed')
else:
    print(f'NOT COMPLETE!')
```

```python
# Double check that each StockCode only has 1 description
assert all_data.groupby(['StockCode']).Description.nunique().max() == 1
```

## 'StockCode' Clean-up

**1. Remove rows where StockCode was for a test item or an adjustment.**

```python
# Initiate list to store rows indices with descriptions starting with 'TEST' or 'ADJUST' 
drop_index = []

# Define regex pattern for finding StockCodes that start with 'TEST' or 'ADJUST'
desc_pattern = re.compile('^(TEST|ADJUST)') 

# For each row in dataframe, check if regex pattern is found and store index to drop_index if it is
for i in range(0,all_data.shape[0]):
    try:
        result = desc_pattern.match(all_data.StockCode.iloc[i])
        if result:
            drop_index.append(i)
    except: 
        if type(all_data.StockCode.iloc[i]) == int:
            continue
```

```python
# Visualize drop_index list
print([x for x in drop_index])
```

```python
# Drop rows with StockCodes for test item or adjustment
all_data = all_data.drop(index = drop_index)
```

## Eliminate Returns for Undocumented Sales

**Returns that are not preceeded by sales that account for the return quantity will be excluded to prevent negative total sales quantities** 

**Generate list of tuples representing all returns**

```python
# Set up mask
return_mask = all_data[all_data['Type']== 'return'].groupby(['CustomerID', 'StockCode', 'InvoiceDate']).sum().reset_index()
check_list = []
return_mask

# for each stockcode customer combo with a return
# find all sales for that combo occuring before date of last retrun
# if sum of all sales qty < sum of return qty then drop return
for row in range(0,return_mask.shape[0]):
    check_list.append((return_mask.loc[row,'CustomerID'],
                       return_mask.loc[row,'StockCode'],
                       return_mask.loc[row,'InvoiceDate']))
```

**Assemble list of indices to drop**

```python
index_to_drop = []
count = 0

# For each item in the checklist, add its index to index_to_drop list if 
for item in check_list:
    
    # assign variables to tuple items
    custID = item[0]
    stockCo = item[1]
    rowDate = item[2]

    # create mask for that customer and StockCode where transaction is a return
    cust_mask = (all_data[(all_data['CustomerID'] == custID) & 
                          (all_data['StockCode'] == stockCo) & 
                          (all_data['Type'] == 'return')])
    
    # count number of unique InvoiceDates
    n_dates = cust_mask.InvoiceDate.nunique()
    
    # list unique InvoiceDates
    unique_dates = cust_mask.InvoiceDate.unique()
    
    # sum all returns for specific StockCode and CustomerID
    return_qty = cust_mask.Quantity.sum()
    
    # ID most recent return of specific product
    return_date = max(unique_dates)
    
    # create a mask to show all purchases by the customer of this specific product before the date of the last return
    sale_mask = (all_data[(all_data['CustomerID'] == custID) & 
                               (all_data['StockCode'] == stockCo) & 
                               (all_data['InvoiceDate'] <= return_date) &
                               (all_data['Type'] == 'sale')])
    
    # sum all purchases of the specific item up to the most recent return date
    buy_qty_to_date = sale_mask.Quantity.sum()

    # subtract qty returned (negative number) from qty purchased to get net_sale qty for the item/customer pair
    net_sale = buy_qty_to_date + return_qty
    
    # if the return date = the most recent return of this item and the qty returned is more than what has been purchased in record
    # add all invoice dates for that item/customer pair to index_to_drop list
    # add all sales dates for that item/customer pair to index_to_drop list
    if rowDate == return_date and net_sale < 0:
        return_idx = cust_mask[cust_mask['InvoiceDate'] <= return_date].index
        sale_idx = sale_mask[sale_mask['InvoiceDate'] <= return_date].index
        
        index_to_drop.extend(return_idx)
        index_to_drop.extend(sale_idx)
```

```python
# Number of indices to drop
len(index_to_drop)
```

**1. Drop CustomerID/StockCode pairs that have returns and sales that add up to < 0 (returns from previous time windows)**

```python
# Drop rows corresponding to returns and sales that would yield net qty < 0
all_data = all_data.drop(index = index_to_drop).reset_index(drop = True)
```
